# Introduction

Wild mushrooms are prized for their unique, savory flavor. Some species, however, are poisonous with some being lethally poisonous. The elusiveness of prized edible mushrooms and the danger inherent with their harvest means some fetch prices upwards of $40/lb in the grocery store.

The lack of universal "rules of thumb" for edibility makes safe mushroom harvest difficult. Foragers are recommended to identify a specimen down to the species level using multiple sources before consuming it. 

This project takes a machine learning approach to solve the problem of identifying edible mushrooms. Around three people die in the United States from mushroom poisoning every year so better identification tools could potentially save lives. To achieve this goal first trends and clusters will be identified using exploratory data analysis. Next, the dimensionality of the dataset will be reduced using unsupervised models to avoid the curse of dimensionality. Unsupervised models will also be used to classify mushrooms as edible or poisonous. Finally, the results from the unsupervised model will be compared with results from a supervised learning model.

# Data Cleaning
All of the features in the dataset are categorical. These will need to be converted to numerical features before training my model. For the features where the only values are True and False a simple conversion to binary will suffice. For the features with greater than two possible values either one-hot encoding or label encoding will be used. One-hot encoding is better than label encoding for the features where there is no ordinal relationship between the values. The exceptions to this are ring-number and gill-spacing for which label encoding will be used.

```python
# Load data from csv
df = pd.read_csv('1987_data_no_miss.csv', delimiter=';')

# Convert edibility label to binary values and rename column to be more intuitive
edibility_mapping = {'p': 0, 'e':1}
df['edible'] = df['class'].map(edibility_mapping)
df.drop(['class'], axis=1, inplace=True)

bruises_mapping = {'t': 1, 'f': 0}
df['bruises'] = df['bruises'].map(bruises_mapping)

# crowded (w), close (c), and distant (d) represent gill spacing in ascending order
gill_spacing_mapping = {'w': 0, 'c': 1, 'd': 2}
df['gill-spacing'] = df['gill-spacing'].map(gill_spacing_mapping)

gill_size_mapping = {'b': 1, 'n': 0}
df['gill-size'] = df['gill-size'].map(gill_size_mapping)

# rename stalk-shape so the binary column has a more intuitive meaning
stalk_shape_mapping = {'t': 0, 'e': 1}
df['stalk-enlarging'] = df['stalk-shape'].map(stalk_shape_mapping)
df.drop(['stalk-shape'], axis=1, inplace=True)

# All the observations had the same value for veil-type: partial veil, so this feature can be dropped
df.drop(['veil-type'], axis=1, inplace=True)

ring_number_mapping = {'n': 0, 'o': 1, 't': 2}
df['ring-number'] = df['ring-number'].map(ring_number_mapping)

columns_to_encode = ['cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-attachment', 'gill-color', 'stalk-root', \
                     'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring',\
                     'veil-color', 'ring-type', 'spore-print-color', 'population', 'habitat']
one_hot_encoded_df = pd.get_dummies(df, columns=columns_to_encode)
```

Data cleaning was relatively simple for this dataset. The dataset had no missing values so no imputation was necessary. The categorical nature of all the features meant there weren't any high-leverage outliers. The main challenge came in relating specific mushroom morphology terms to numerical values. For example gill-spacing takes on values crowded, close, and distant. Additional research was required to understand that crowded gills are closer together than close gills. 

One of the drawbacks of using one-hot encoding for taxonomic data is that the resulting distance between two observations is dependent on the number of states that have been used. For example, if cap color is divided into many subtly different shades using a color chart there will be a larger number of mismatches between observations than if it is divided into a few.[5] For instance in this dataset gill-color has 12 possible values while veil-color only has three.

# Dimensionality Reduction
After converting all the dataset to numerical data we are left with 111 features. Due to the curse of dimensionality it is hard to fit data to high dimensional space. 

## Principle Component Analysis
Using Principle Component Analysis I will reduce the dimensionality while preserving the majority of variance. 

```python
# Separate features and labels for dimensionality reduction
y = one_hot_encoded_df['edible']
X = one_hot_encoded_df.drop(['edible'], axis=1).to_numpy()
total_variance = np.var(one_hot_encoded_df, axis=0).sum()

n_component_values = range(1, 109)
explained_variances = []
explained_variances_cum_sum = []
explained_variance = 0
for n_components in n_component_values:
    # currently not scaled
    pca = PCA(n_components=n_components)
    reduced_data = pca.fit_transform(X)
    new_explained_variance = pca.explained_variance_.sum()
    explained_variances_cum_sum.append(new_explained_variance / total_variance)
    explained_variance_ratio = (new_explained_variance - explained_variance) / total_variance
    explained_variances.append(explained_variance_ratio)
    explained_variance = new_explained_variance
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(n_component_values, explained_variances, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Explained Variance vs Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.xticks(range(1, 109, 5))  # Set x-ticks to be the number of components
plt.grid()

# Show the plot
plt.show()
```

There is an elbow in the scree plot at about 7 components but no clear dropoff in proportion of variance explained by each new additional component. Since there are so many components in the initial data set the cumulative effect of components in the tail portion of the skree plot is significant. This can be seen in the plot of cumulative variance explained below. Proportion of explained variance doesn't cross 0.90 until 36 components. Number of components can be tuned as a hyperparameter in cross validation to determine the ideal number

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(n_component_values, explained_variances_cum_sum, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Cumulative Explained Variance vs Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.xticks(range(1, 109, 5))  # Set x-ticks to be the number of components
plt.grid()

# Show the plot
plt.show()
```

## Multiple Correspondence Analysis
The goal of principle component analysis is reduce the dimensionality of the dataset using vectors that are as highly correlated as possible with the data vectors. It is possible to interpret the correlation between a discrete variable (like our original dataset) and continous variable (what is required to train our model) if the continouous variable is approximated as using an additive effects model based on the observed character states of the discrete variable. [4] Multiple Correspondence Analysis uses this interpretation to extend Principle Component Analysis to categorical data.

```python
df = pd.read_csv('1987_data_no_miss.csv', delimiter=';')
df_only_features = df.drop(['class'], axis=1)
mca = MCA(
    n_components=82,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)
mca = mca.fit(df_only_features)
summary = mca.eigenvalues_summary
print(summary)
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(summary.index, summary['% of variance (cumulative)'], marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Cumulative Explained Variance vs Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.yticks(ticks=range(0, 100, 10))
plt.grid()

# Show the plot
plt.show()
```

Interestingly, the resulting eigenvalues for Multiple Correspondence Analysis show an equal proportion of variance explained all the way to 100% at 82 components. I will compare the results of using components from MCA with components from PCA in the supervised model.

# Analysis Using Unsupervised Model

Since there are two labels, edible and poisonous, in the original dataset I would like to see how clustering algorithms divide the mushroom species into two groups. One of the disadvantages of k-means compared with hierarchical clustering is that you need to specify the number of clusters beforehand. Since I already know I want two clusters I will use k-means. Since k-means is a greedy algorithm it finds a local optimum which minimizes within-cluster variation. To find the optimal solution the algorithm needs to be run multiple times by adjusting the n_init parameter.

```python
# Train kmeans clustering model and predict edibility labels
kmeans = KMeans(n_clusters=2, n_init=5)
predicted_labels = kmeans.fit_predict(X)

# clustering resulted in swapped labels so swap them back
yp = [0 if label == 1 else 1 for label in predicted_labels]

# Evaluate the model using different metrics
acc = accuracy_score(y, predicted_labels)
print(f'K-means clustering accuracy: {acc}')

f1 = f1_score(y, predicted_labels)
print(f'K-means clustering F1 score: {f1}')

# Confusion Matrix
cm = confusion_matrix(y, predicted_labels)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True, 
            xticklabels=['Predicted poisonous', 'Predicted edible'], 
            yticklabels=['True poisonous', 'True edible'])

# Add titles and labels
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Show the plot
plt.show()
```

The k-means algorithm performs reasonably well with an accuracy of 0.90 and F1 score of 0.91. As can be seen from the confusion matrix, however, most of the incorrectly predicted labels were poisonous mushrooms that were predicted to be edible. Two disadvantages with k-means clustering could lead to this lower accuracy score. K-means assumes a spherical structure to the clusters which may not reflect the actual structure of the mushroom data. Clustering algorithms also do not use the labels as ground truth to guide the classification process meaning the resulting clusters are not based on the actual edible/poisonous split. 

If the goal of this project is to make mushroom foraging safer through effective identification the false positive rate needs to be lower. Next I will use compare the results from k-means to a supervised model: Support Vector Machine.

# Analysis Using Supervised Learning
I chose Support Vector Machine because it is effective in high-dimensional spaces. Even after dimensionality reduction this dataset has a lot of features. SVM is also robust to overfitting. 

Since there is no dedicated test dataset I will use k-fold cross validation to estimate test error. I will choose a grid of hyperparameters, calculate cross-validation error for each, then use the hyperparameters that result in the lowest estimated test error for my final model. The hyperparameters I will tune for this project are num_components in the dimensionality reduction phase and the regularization paramater for SVM. I will also compare dimensionality reduction using MCA with PCA.

I initially tried number of components from 1 to 81 in increments of 5 in order to reach a proportion of variance explained of 100%. All the best performing models used 1-10 components so I reduced the range to 1 to 20 in increments of 1 for a more granular view.


```python
# start at minimum number of components and go up to near 100% proportion of variance explained
num_components_options = range(1, 20)
param_grid = {'C': [0.1, 1, 10, 100]}
svc = SVC(kernel='rbf')
X_df = one_hot_encoded_df.drop(['edible'], axis=1)
results_pca = []
results_mca = []

for num_components in num_components_options:
    pca = PCA(n_components=num_components, random_state=42)
    pca_reduced_data = pca.fit_transform(X)

    mca = MCA(n_components=num_components, one_hot=False, random_state=42)
    mca.fit(X_df)
    mca_reduced_data = mca.transform(X_df)

    clf_pca = GridSearchCV(estimator=svc, param_grid=param_grid, scoring='f1')
    clf_pca.fit(pca_reduced_data, y)
    results_pca.append(clf_pca.cv_results_)

    clf_mca = GridSearchCV(estimator=svc, param_grid=param_grid, scoring='f1')
    clf_mca.fit(mca_reduced_data, y)
    results_mca.append(clf_mca.cv_results_)
    
print(results_mca[0])
```

```python
pca_test_scores = [d['mean_test_score'] for d in results_pca]
columns = [0.1, 1, 10, 100]
pca_results_df = pd.DataFrame(pca_test_scores, columns=columns)
pca_results_df.index = range(1, 20)

# Plot results of hyperparameter tuning for PCA components
plt.figure(figsize=(10, 6))
for col in pca_results_df.columns:
    plt.plot(pca_results_df.index, pca_results_df[col], marker='o', label=f'C: {col}')

plt.title('Mean F1 Scores by Number of PCA Components')
plt.xlabel('Components')
plt.ylabel('Mean F1 Score')
plt.xticks(ticks=range(1, 20))  # Set x-ticks to be the row indices
plt.grid()
plt.legend(title='Columns')
plt.axis([0, 20, 0.5, 1])

plt.show()
```

```python
mca_test_scores = [d['mean_test_score'] for d in results_mca]
mca_results_df = pd.DataFrame(mca_test_scores, columns=columns)
mca_results_df.index = range(1, 20)

# Plot results of hyperparameter tuning for MCA components
plt.figure(figsize=(10, 6))
for col in mca_results_df.columns:
    plt.plot(mca_results_df.index, mca_results_df[col], marker='o', label=f'C: {col}')

plt.title('Mean F1 Scores by Number of MCA Components')
plt.xlabel('Components')
plt.ylabel('Mean F1 Score')
plt.xticks(ticks=range(1, 20))  # Set x-ticks to be the row indices
plt.grid()
plt.legend(title='Columns')
plt.axis([0, 20, 0.5, 1])

plt.show()
```

## Hyperparameter Tuning Analysis
The regularization parameter C in sklearn is defined as the inverse of the textbook definition. A C value of 0.1 has the widest margin for misclassifications when creating the hyperplane which leads to higher bias as seen in the resulting low F1 scores. The highest F1 score corresponded to a C value of 100 which generally corresponds to lower bias and higher variance than 0.1. One of the dangers of automatically selecting the C parameter is overfitting. While cross validation estimates the test error I would be curious to see how the best model (5 PCA components and C value of 100) performs on unseen data.

There is no significant difference in F1 score between components generated from Multiple Correspondence Analysis and components from Principle Component Analysis. Even though MCA is designed for categorical data the best performing model used components generated using PCA. I one-hot encoded the data before dimensionality reduction rather than relying on MCA's built in encoding which may have caused the two methods to perform similarly. If all the features are dichotomous (yes/no) MCA and PCA become essentially identical. Since some of the features in this dataset are dichotomous and none have more than 12 potential values this could explain some of the similarity between the two methods.

```python
# Final Model

# Dimensionality reduction
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.25)
pca = PCA(n_components=5, random_state=42).fit(X_train)
X_train_reduced = pca.transform(X_train)
X_test_reduced = pca.transform(X_test)

# Train the model and generate predictions
svc = SVC(kernel='rbf', C=100)
svc = svc.fit(X_train_reduced, y_train)
y_pred_train = svc.predict(X_train_reduced)
y_pred_test = svc.predict(X_test_reduced)

# Accuracy
train_acc = accuracy_score(y_train, y_pred_train)
test_acc = accuracy_score(y_test, y_pred_test)
print(f'Training accuracy: {train_acc}')
print(f'Test accuracy: {test_acc}')

# F1 Score
train_f1 = f1_score(y_train, y_pred_train)
test_f1 = f1_score(y_test, y_pred_test)
print(f'Training F1 score: {train_f1}')
print(f'Test F1 score: {test_f1}')

# ROC AUC
train_roc = roc_auc_score(y_train, y_pred_train)
test_roc = roc_auc_score(y_test, y_pred_test)
print(f'Training ROC-AUC score: {train_roc}')
print(f'Test ROC-AUC score: {test_roc}')

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_test)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True, 
            xticklabels=['Predicted poisonous', 'Predicted edible'], 
            yticklabels=['True poisonous', 'True edible'])

# Add titles and labels
plt.title('Test Data Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

# Show the plot
plt.show()
```

# Discussion and Conclusions

To evaluate the model using the tuned hyperparameters I split the data into a training portion and test portion. The model performed well at predicting the edibility of mushrooms in both the training set and the test set. The training accuracy for SVM was 0.998 and the test accuracy was 0.999. The other performance metrics can be seen above.

The supervised model performed better than the unsupervised model at classifying the edibility of mushrooms. This superior performance is expected since the supervised model leverages the labels in the training set to guide the classification process.

One potential way to improve the performance of the model would be to select non-consecutive principle components. There are jumps in F1 score at 2, 5, 7, and 14 components. A model using exclusively these components may perform better than one using all 14 components with the highest variance explained. Another way to improve the performance of the model would be to center and normalize the data before doing dimensionality reduction.
